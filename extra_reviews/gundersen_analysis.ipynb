```python
# Problems IDs encoding

problem_ids = {  # Mapping of the problem definition of Gundersen et al. to our method
    1: "None",  # Method code is shared, but not experiment code. 
    2: "Implementation",  # Method code is shared but does not cover the entire method. 
    3: "Implementation",  # Poor documentation of code. 
    4: "Experimental Procedure",  # Experiment setup not not described or differ from article.
    5: "Implementation",  # Not clear which version of the code is used for the experiments. 
    6: "Implementation",  # Code is shared in a compiled or non-inspectable form.
    7: "None",  # Random seeds and random number generators not specified. 
    8: "Implementation",  # Ambiguous description of method.
    9: "Experimental Procedure",  # Ambiguous description of the experiment. 
    10: "Implementation",  # Ambiguous description of the implementation of the method or experiment.
    11: "Implementation",  # Multiple methods for solving a sub-task are mentioned but not specified
    12: "Configuration",  # Hyperparameters are shared online, but differ from those used in experiment.
    13: "Configuration",  # Experiment- or hyper-parameters are not shared.
    14: "None",  # The article contains an error.
    15: "Data",  # Mismatch between dataset specified and version of it found online.
    16: "Experimental Procedure",  # A subset of a dataset is used, but which exactly is not specified. 
    17: "Data",  # The preprocessed or augmented version of a dataset is not shared. 
    18: "Experimental Procedure",  # How dataset is partitioned into training, validation, and test is not described.
    19: "None",  # Results are presented in a way that makes a comparison hard. 
    20: "None",  # Lack of access to hardware or software needed to conduct the experiment. 
}

problem_ids_occurences = {
    1: 5,
    2: 1,
    3: 1,
    4: 2,
    5: 3,
    6: 2,
    7: 4,
    8: 5,
    9: 7,
    10: 12,
    11: 2,
    12: 1,
    13: 6,
    14: 1,
    15: 5,
    16: 1,
    17: 4,
    18: 4,
    19: 2,
    20: 1,
}
```

```python
# Extract reviews data
from pathlib import Path
import pandas as pd

review_df = []
for review in Path.cwd().rglob("*.md"):
    implementation, data, config, exp = None, None, None, None
    review_lines = review.open().readlines()
    for index, line in enumerate(review_lines):
        if not implementation and "### Implementation" in line:
            implementation = int(review_lines[index + 3].strip()[1:-1])
        if not data and "### Data" in line:
            data = int(review_lines[index + 3].strip()[1:-1])
        if not config and "### Configuration" in line:
            config = int(review_lines[index + 3].strip()[1:-1])
        if not exp and "### Experimental Procedure" in line:
            exp = int(review_lines[index + 3].strip()[1:-1])
        if implementation and data and config and exp:
            break
    #review_df.append([review.stem, implementation, data, config, exp])
    gundersen_results[review.stem].update({"Implementation": implementation, "Data": data, "Configuration": config, "Experimental Procedure": exp})

#review_df = pd.DataFrame(review_df, columns=["Review", "Implementation", "Data", "Configuration", "Experimental Procedure"])
df = pd.DataFrame(gundersen_results.values(), index=gundersen_results.keys())
df["n_experiments_success"] = df["n_results_identical"] + df["n_results_consistent"]
```

```python
for index, row in df.iterrows():
    if (row["result"] == "F" or row["result"] == "NR") and row["time"] < 40:
        df["time"][index] = 40  # Penalty

result_map = {
    "S": True,
    "PS": True,
    "NR": False,
    "F": False,
}

df["success"] = df["result"].map(result_map)
for column in ["n_experiments_completed", "n_results_identical", "n_results_consistent", "n_experiments_success", "n_experiments_failed"]:
    if df[column].dtype == int:  # Still count, convert to percentage
        df[column] = df[column] / df["n_experiments"]

```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def scatter_with_jitter(df, x_axis="Implementation", y_axis="Configuration", hue="n_experiments_success", point_label="Experimental Procedure"):

    # ----- Color map (bad → medium → good) -----
    cmap = ListedColormap([
        "#8B0000",  # dark red
        "#FF4500",  # orange-red
        "#FFD700",  # yellow
        "#9ACD32",  # yellow-green
        "#006400"   # dark green
    ])

    df = df.copy()

    # Group duplicates
    groups = df.groupby([x_axis, y_axis])

    # Prepare jittered coordinates
    jitter_x = []
    jitter_y = []

    jitter_radius = 0.25  # adjust if you want more/less spacing

    for (impl, conf), group in groups:
        n = len(group)

        # If only one point → no jitter
        if n == 1:
            jitter_x.append([impl])
            jitter_y.append([conf])
            continue

        # For duplicates → spread them in a circle
        angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
        jitter_x.append(impl + np.cos(angles) * jitter_radius)
        jitter_y.append(conf + np.sin(angles) * jitter_radius)

    # Flatten lists back into series
    df["jitter_x"] = np.concatenate(jitter_x)
    df["jitter_y"] = np.concatenate(jitter_y)

    # ----- Plot -----
    plt.figure(figsize=(9, 7))

    sc = plt.scatter(
        df["jitter_x"],
        df["jitter_y"],
        c=df[hue],
        cmap=cmap,
        s=110,
        edgecolor="black"
    )

    # Labels (Experimental Procedure)
    for _, row in df.iterrows():
        plt.text(
            row["jitter_x"] + 0.03,
            row["jitter_y"] + 0.03,
            f"{row[point_label]:.2f}",
            fontsize=9,
            ha="left",
            va="bottom"
        )

    # Colorbar
    cbar = plt.colorbar(sc)
    cbar.set_label(hue)

    # Axis formatting
    plt.xlabel("Implementation")
    plt.ylabel("Configuration")
    plt.xticks(range(1, 11))
    plt.yticks(range(1, 11))
    plt.grid(True, linestyle="--", alpha=0.3)

    plt.title("Scatter Plot with Discrete Color Map, Non-Overlapping Jitter, and Labels")
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.genmod.generalized_linear_model import GLM
from statsmodels.genmod import families

# --- 3. Define Predictor and Target Variables ---
X = df[dimensions]
y = df['n_experiments_success']

# --- 4. Add a Constant (Intercept) to the Predictors ---
# statsmodels requires an explicit constant term for the intercept
X = sm.add_constant(X, prepend=False)

# --- 5. Estimate the Probit Model ---
# We specify:
# a) family=sm.families.Binomial() for binary outcomes (0 or 1)
# b) link=sm.families.links.Probit() to use the Probit link function
probit_model = GLM(y, X, family=families.Binomial(link=families.links.Probit()))
probit_results = probit_model.fit()

# --- 6. Print the Results Summary ---
print("### Probit Regression Results (statsmodels) ###")
print(probit_results.summary())
```

```python
# Calculate and print Marginal Effects
print("\n" + "="*50 + "\n")
print("### Average Marginal Effects (AMEs) ###")
# AMEs are the average of the marginal effect calculated for each observation
print(probit_results.get_margeff(at='overall').summary())
```
